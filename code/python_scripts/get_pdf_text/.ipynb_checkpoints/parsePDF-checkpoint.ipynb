```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        
        # Iterate through each page and extract text
        text = ''
        for page in pdf_reader.pages:
            text += page.extract_text() + '\n'  # Extract text from each page
        
    return text

# Specify the path to your PDF
pdf_path = 'CTCI_7_12.pdf'
text = extract_text_from_pdf(pdf_path)
print(text)
```

```python
from PyPDF2 import PdfReader, PdfWriter

def extract_pages(source_pdf_path, output_pdf_path, start_page, end_page):
    # Create a PDF reader object
    reader = PdfReader(source_pdf_path)
    # Create a PDF writer object for the output
    writer = PdfWriter()

    # Adding pages from start_page to end_page to the writer object
    for i in range(start_page - 1, end_page):  # Page numbers are zero-indexed in PyPDF2
        writer.add_page(reader.pages[i])

    # Write the pages to a new file
    with open(output_pdf_path, 'wb') as output_pdf:
        writer.write(output_pdf)
    print(f"Pages from {start_page} to {end_page} have been extracted to {output_pdf_path}")

# Specify the path to your PDF and the output path
source_pdf_path = 'CTCI.pdf'
output_pdf_path = 'CTCI_7_12.pdf'

# Call the function with the range of pages you want to extract
extract_pages(source_pdf_path, output_pdf_path, 7, 12)
```
